```python
from openai import OpenAI
```

```python
openai = OpenAI()
```

```python
messages = [{"role": "user", "content": "what is 2+2?"}]
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

print(response.choices[0].message.content)
```

```python
question = "Please propose a hard challenging question to assess someone's IQ. Respond only with the question."
messages = [{"role": "user", "content": question}]

```

```python
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

question = response.choices[0].message.content

print(question)
```

```python
messages = [{"role": "user", "content": question}]
```

```python
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

answer = response.choices[0].message.content

print(answer)
```

```python
from IPython.display import Markdown, display

display(Markdown(answer))
```

```python
#Excercise 1

messages = [{"role": "user", "content": "Pick a business area that might be worth exploring for agentic ai oppotunity?"}]
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

business_area = response.choices[0].message.content

print(business_area)

```

```python
display(Markdown(business_area))
```

```python
messages = [{"role": "user", "content": "Please tell me what are the pain points of the business area " + business_area + "? Something challenging that might be ripe for agentic solutions?"}]

response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

pain_points = response.choices[0].message.content

print(pain_points)

```

```python
display(Markdown(pain_points))
```

```python
messages = [{"role": "user", "content": "Can you propose a solution to the problem " + pain_points + "? Respond only with the solution."}]

response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

solution = response.choices[0].message.content

print(solution)

display(Markdown(solution))
```
