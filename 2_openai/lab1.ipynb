```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
```

```python
load_dotenv(override=True)
```

```python
# make an agent with name, instructions, model
# instructions are basically the sys prompt, to this llm

# you can think of each of these agents as llm with one particular sys prompt designed to oraganize around one 
# paricular task.

agent = Agent(name="Jokester", instructions="You are a joke teller", model="gpt-4o-mini")
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

result = Runner.run(agent, "Tell a joke about autonomous AI agents")
print(result)

# output is a coroutine obj, bcoz runner.run is an async method
```

```python
result = await Runner.run(agent, "Tell a joke about autonomous AI agents")
print(result.final_output)

# no need to run this the same is run within trace
# what trace does is, it tells openai to record these agent interactions.
```

```python
with trace("telling a joke"):
    result = await Runner.run(agent, "Tell a joke about autonomous AI agents")
    print(result.final_output)
```
